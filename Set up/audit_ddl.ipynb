```python
%sql

CREATE SCHEMA IF NOT EXISTS tt_hc_adb_ws.audit;

CREATE TABLE IF NOT EXISTS tt_hc_adb_ws.audit.load_logs (
    data_source STRING,
    tablename STRING,
    numberofrowscopied INT,
    watermarkcolumnname STRING,
    loaddate TIMESTAMP
);
```

```python
dbutils.widgets.text("data_source", "")
dbutils.widgets.text("tablename", "")
dbutils.widgets.text("numberofrowscopied", "")
dbutils.widgets.text("watermarkcolumnname", "")
dbutils.widgets.text("loaddate", "")
```

```python
data_source = dbutils.widgets.get("data_source")
table_name = dbutils.widgets.get("tablename")
rows_copied = int(dbutils.widgets.get("numberofrowscopied"))
watermark_column = dbutils.widgets.get("watermarkcolumnname")
load_date = dbutils.widgets.get("loaddate")
```

```python
spark.sql(f"""
INSERT INTO tt_hc_adb_ws.audit.load_logs
(
data_source,
tablename,
numberofrowscopied,
watermarkcolumnname,
loaddate
)
VALUES
(
'{data_source}',
'{table_name}',
'{rows_copied}',
'{watermark_column}',
TIMESTAMP('{load_date}')
)
""")
```
