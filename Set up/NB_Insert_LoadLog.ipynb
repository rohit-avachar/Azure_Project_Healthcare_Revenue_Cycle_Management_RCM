```python
# =====================================================
# Notebook Name : NB_Insert_LoadLog
# Purpose       : Insert Audit Log into Delta Table
# =====================================================

# Create widgets
dbutils.widgets.text("data_source", "")
dbutils.widgets.text("tablename", "")
dbutils.widgets.text("numberofrowscopied", "")
dbutils.widgets.text("watermarkcolumnname", "")
dbutils.widgets.text("loaddate", "")

# Read parameters
data_source = dbutils.widgets.get("data_source")
tablename = dbutils.widgets.get("tablename")
numberofrowscopied = int(dbutils.widgets.get("numberofrowscopied"))
watermarkcolumnname = dbutils.widgets.get("watermarkcolumnname")
loaddate = dbutils.widgets.get("loaddate")

# print(f"Data Source      : {data_source}")
# print(f"Table Name       : {tablename}")
# print(f"Rows Copied      : {rows_copied}")
# print(f"Watermark Column : {watermark_column}")
# print(f"Load Date        : {load_date}")

query = f"""
INSERT INTO tt_hc_adb_ws.audit.load_logs
(
    data_source,
    tablename,
    numberofrowscopied,
    watermarkcolumnname,
    loaddate
)
VALUES
(
    '{data_source}',
    '{tablename}',
    {numberofrowscopied},
    '{watermarkcolumnname}',
    TIMESTAMP('{loaddate}')
)
"""

print("Executing SQL:")
print(query)

spark.sql(query)

print("Audit record inserted successfully.")

dbutils.notebook.exit("SUCCESS")
```
