```python
%sql
CREATE TABLE IF NOT EXISTS tt_hc_adb_ws.gold.dim_icd (
    icd_code STRING,
    icd_code_type STRING,
    code_description STRING,
    refreshed_at TIMESTAMP
)
```

```python
%sql
truncate table tt_hc_adb_ws.gold.dim_icd
```

```python
%sql
insert into tt_hc_adb_ws.gold.dim_icd
select distinct
  icd_code,
  icd_code_type,
  code_description,
  current_timestamp() refreshed_at
from
  tt_hc_adb_ws.silver.icd_codes
where
  is_current_flag = true
```
