```python
%sql
insert into tt_hc_adb_ws.gold.dim_patient
select 
     patient_key ,
    src_patientid ,
    firstname ,
    lastname ,
    middlename ,
    ssn ,
    phonenumber ,
    gender ,
    dob ,
    address ,
    datasource 
 from tt_hc_adb_ws.silver.patients
 where is_current=true and is_quarantined=false
```
